Calcular la resistencia equivalente

```python
import pandas as pd
```

```python
# Se crea la funcion para realizar la sumatoria

def serie(valores:list):
    Req = pd.DataFrame(valores)
    return Req.sum()
```

```python
# Se llama la función
Rtotal = serie([100,200,220,140,220])
print(int(Rtotal),"Ohms")
```

```python
# Funcion para calcular la corriente
def corriente(voltaje,resistencia):
    i = voltaje/resistencia
    return i
```

```python
Itotal = corriente(1,Rtotal)
print(float(round(Itotal,3)),"=> 1mA")
```

```python
# Voltaje en R3
def voltaje(corriente,resistencia):
    return corriente*resistencia
```

```python
VR2=voltaje(Itotal,220)
print(float(VR2),"V")
```
